```python
import gymnasium as gym
from stable_baselines3 import PPO

env = gym.make('MountainCar-v0', render_mode='human')
model = PPO("MlpPolicy", env, verbose=1)
model.learn(total_timesteps=10000)
model.save("ppo_mountaincar")

obs, info = env.reset()  # Gym environment returns a tuple

for _ in range(100):
    action, _states = model.predict(obs)
    obs, reward, done, truncated, info = env.step(action)
    env.render()

    if done:
        obs, _ = env.reset()

```

```python
import gymnasium as gym
from stable_baselines3 import PPO
from stable_baselines3.common.env_util import make_vec_env

# Creating the VecEnv (multiple environments)
vec_env = make_vec_env("MountainCar-v0", n_envs=1)  # Can set n_envs to any number for parallel envs

model = PPO("MlpPolicy", vec_env, verbose=1)
model.learn(total_timesteps=10000)
model.save("ppo_mountaincar_vecenv")

# For VecEnv, reset() only returns the observations, not the info
obs = vec_env.reset()  # Returns only the observations

while True:
    action, _states = model.predict(obs)  # Predictions based on the observations
    obs, reward, done, truncated, info = vec_env.step(action)
    vec_env.render()  # Rendering the environment
```

```python
import gymnasium as gym
from stable_baselines3 import PPO

# Create the MountainCar environment
env = gym.make('MountainCar-v0', render_mode='human')

# Initialize the PPO model
model = PPO("MlpPolicy", env, verbose=1, learning_rate=0.0003)

# Train the model for a longer period
model.learn(total_timesteps=200000)

# Save the trained model
model.save("ppo_mountaincar")

# After training, reset the environment and test the model
obs, info = env.reset()

for i in range(500):  # You can test for a longer period
    action, _states = model.predict(obs)
    obs, reward, done, truncated, info = env.step(action)
    env.render()

    if done:
        break
    print(f" Episodes: {i+1}")
env.close()  # Close the environment after testing
```
